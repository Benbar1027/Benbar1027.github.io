```python
import pandas as pd
import numpy as np

def surname_cleaner(dataframe):
    """
    Cleans and processes a DataFrame containing census data. Specifically, it:
    1. Replaces '(S)' values with NaN.
    2. Converts percentage columns (white, black, API, AIAN, 2+ races, Hispanic) to proportions.
    3. Counts the number of missing values (NaNs) for these demographic columns.
    4. Calculates the remaining proportion for each row.
    5. Distributes this remaining proportion equally among the missing values in each row.
    6. Converts the 'name' column values to lowercase.

    Args:
    dataframe (pd.DataFrame): A DataFrame containing the census data with specific columns.

    Returns:
    pd.DataFrame: A cleaned and processed DataFrame.
    """

    # Replace '(S)' values with NaN
    dataframe = dataframe.replace('(S)', np.nan)

    # Convert percentage columns to proportions, handling NaNs
    for column in ['pctwhite', 'pctblack', 'pctapi', 'pctaian', 'pct2prace', 'pcthispanic']:
        dataframe[column] = pd.to_numeric(dataframe[column], errors='coerce') / 100

    # Count the number of missing values in each row for specified columns
    missing_columns = ['pctwhite', 'pctblack', 'pctapi', 'pctaian', 'pct2prace', 'pcthispanic']
    dataframe['countmiss'] = dataframe[missing_columns].isnull().sum(axis=1)

    # Calculate the remaining proportion for each row
    dataframe['remaining'] = dataframe['count'] * (1 - dataframe[missing_columns].sum(axis=1, min_count=1))

    # Distribute the remaining proportion equally among the missing values
    for column in missing_columns:
        dataframe.loc[dataframe[column].isnull(), column] = dataframe['remaining'] / (dataframe['countmiss'] * dataframe['count'])

    # Convert the name column to lowercase
    dataframe['name'] = dataframe['name'].str.lower()

    return dataframe

```

```python
#BISG Example Smith California 

def calculate_bisg(surname_df, geo_df, surname):
    """
    Calculate the Bayesian Improved Surname Geocoding (BISG) probability.

    Args:
    surname_df (pd.DataFrame): DataFrame containing surname distribution across races.
    geo_df (pd.DataFrame): DataFrame containing geographic distribution across races.
    surname (str): The target surname to calculate probabilities for.

    Returns:
    pd.DataFrame: BISG probabilities for the target surname.
    """
    
    # Ensure the dataframes are indexed by race for easier lookup
    surname_df = surname_df.set_index('Race/Ethnicity')
    geo_df = geo_df.set_index('Race/Ethnicity')

    # Get the distribution for the target surname across different races
    surname_dist = surname_df[surname].to_dict()

    # Calculate the BISG probabilities
    bisg_probs = {}
    total = 0
    for race, p_r_given_s in surname_dist.items():
        # q(g|r) - proportion of race in geographic area
        q_g_given_r = geo_df.at[race, '% of U.S. Population Residing in California']

        # Calculate the numerator p(r|s) * q(g|r)
        bisg_probs[race] = p_r_given_s * q_g_given_r
        total += bisg_probs[race]

    # Normalize the probabilities to sum to 1
    for race in bisg_probs:
        bisg_probs[race] /= total

    # Convert to DataFrame
    bisg_df = pd.DataFrame([bisg_probs], index=[surname])
    return bisg_df

# Example data from the images provided by the user
surname_df = pd.DataFrame({
    'Race/Ethnicity': ['Hispanic', 'White', 'Black', 'Asian and Pacific Islander',
                       'American Indian/Alaska Native', 'Multiracial'],
    'Smith': [0.0156, 0.7335, 0.2222, 0.0040, 0.0085, 0.0163]
})

geo_df = pd.DataFrame({
    'Race/Ethnicity': ['Hispanic', 'White', 'Black', 'Asian and Pacific Islander',
                       'American Indian/Alaska Native', 'Multiracial'],
    '% of U.S. Population Residing in California': [0.2776, 0.0791, 0.0603, 0.3335, 0.0786, 0.1752]
})

# Calculate BISG probabilities for 'Smith'
bisg_result = calculate_bisg(surname_df, geo_df, 'Smith')
bisg_result
```

```python
import pandas as pd
import numpy as np

def calculate_bisfg_probabilities(input_df, block, tract, zip_code_df, first_names, surname_2010_census):
    """
    Calculate the Bayesian Improved Surname Geocoding (BISFG) probabilities for each racial group.

    Parameters:
    input_df (DataFrame): DataFrame containing the columns 'firstname', 'name', and may include 'GeoInd_Block', 'GeoInd_Tract', 'ZCTA5'.
    block (DataFrame): DataFrame containing block-level geographical data.
    tract (DataFrame): DataFrame containing tract-level geographical data.
    zip_code_df (DataFrame): DataFrame containing ZIP code-level geographical data.
    first_names (DataFrame): DataFrame containing probabilities based on first names, including a row for 'ALL OTHER FIRST NAMES'.
    surname_2010_census (DataFrame): DataFrame containing probabilities based on surnames, including a row for 'all other names'.

    Returns:
    DataFrame: A DataFrame with the BISFG probabilities for each racial group.
    """
    
    # Standardize the 'firstname' and 'name' columns for consistency in matching
    input_df['firstname'] = input_df['firstname'].str.upper()
    input_df['name'] = input_df['name'].str.lower().str.replace('-', '_')

    # Copy the input DataFrame to keep the original data intact
    output_df = input_df.copy()

    # Define the suffixes for the racial groups in the surname probability dataset
    race_suffixes = {
        'white': 'white', 'black': 'black', 'aian': 'aian',
        'api': 'api', 'mult_other': '2prace', 'hispanic': 'hispanic'
    }

    # Define the function to calculate the BISFG probability for a row
    def bisfg_probability(row, race_key, race_suffix):
        p_r_s = p_f_r = p_g_r = 1

        # Handle hyphenated surnames by splitting and checking each part
        surnames = row.get('name', '').split('_')
        found_surname = False
        for surname in surnames:
            # Check if the surname is in the surname_2010_census dataframe
            matching_surnames = surname_2010_census[surname_2010_census['name'] == surname]
            if not matching_surnames.empty:
                p_r_s = matching_surnames[f'pct{race_suffix}'].iloc[0]
                found_surname = True
                break

        # Use 'all other names' probabilities if surname is not found
        if not found_surname:
            p_r_s = surname_2010_census[surname_2010_census['name'] == 'all other names'][f'pct{race_suffix}'].iloc[0]

        # Check for the first name in the first names dataset and get its probability
        firstname = row.get('firstname', '')
        matching_firstnames = first_names[first_names['firstname'] == firstname]
        if not matching_firstnames.empty:
            p_f_r = matching_firstnames[f'name_given_{race_key}'].iloc[0]
        else:
            # Use 'ALL OTHER FIRST NAMES' probabilities if firstname is not found
            p_f_r = first_names[first_names['firstname'] == 'ALL OTHER FIRST NAMES'][f'name_given_{race_key}'].iloc[0]

        # Determine the geography-based probability using the block, tract, or ZIP code data
        if 'GeoInd_Block' in input_df.columns and row.get('GeoInd_Block') and row['GeoInd_Block'] in block['GeoInd_Block'].values:
            p_g_r = block[block['GeoInd_Block'] == row['GeoInd_Block']][f'here_given_{race_key}'].iloc[0]
        elif 'GeoInd_Tract' in input_df.columns and row.get('GeoInd_Tract') and row['GeoInd_Tract'] in tract['GeoInd_Tract'].values:
            p_g_r = tract[tract['GeoInd_Tract'] == row['GeoInd_Tract']][f'here_given_{race_key}'].iloc[0]
        elif 'ZCTA5' in input_df.columns and row.get('ZCTA5') and row['ZCTA5'] in zip_code_df['ZCTA5'].values:
            p_g_r = zip_code_df[zip_code_df['ZCTA5'] == row['ZCTA5']][f'here_given_{race_key}'].iloc[0]
        else:
            p_g_r = 1

        # Calculate the numerator of the BISFG probability
        numerator = p_r_s * p_f_r * p_g_r
        
        # Calculate the denominator as the sum of all possible combinations of surname, firstname, and geography probabilities
        denominator = sum(
            p_surname * p_firstname * p_geography
            for key, suffix in race_suffixes.items()
            for part in surname.split('_')
            for geo_df, geo_col in [(block, 'GeoInd_Block'), (tract, 'GeoInd_Tract'), (zip_code_df, 'ZCTA5')]
            if part in surname_2010_census['name'].values
            for p_surname in [surname_2010_census[surname_2010_census['name'] == part][f'pct{suffix}'].iloc[0]]
            if firstname in first_names['firstname'].values
            for p_firstname in [first_names[first_names['firstname'] == firstname][f'name_given_{key}'].iloc[0]]
            if row.get(geo_col) in geo_df[geo_col].values
            for p_geography in [geo_df[geo_df[geo_col] == row.get(geo_col)][f'here_given_{key}'].iloc[0]]
        )


        # Print numerator and denominator
        #print(f"Numerator for {race_key}: {numerator}")
        #print(f"Denominator for {race_key}: {denominator}")

        # Return the BISFG probability for the row, handling division by zero
        return numerator / denominator if denominator != 0 else np.nan
       
    # Calculate the BISFG probabilities for each racial group and add them to the output DataFrame
    for key, suffix in race_suffixes.items():
        output_df[f'bisfg_{key}'] = input_df.apply(lambda row: bisfg_probability(row, key, suffix), axis=1)

    return output_df


```

```python
import pandas as pd
import numpy as np

def calculate_bisfg_probabilities(input_df, block, tract, zip_code_df, first_names, surname_2010_census):
    """
    Calculate the Bayesian Improved Surname Geocoding (BISFG) probabilities for each racial group.

    Parameters:
    input_df (DataFrame): DataFrame containing the columns 'firstname', 'name', and may include 'GeoInd_Block', 'GeoInd_Tract', 'ZCTA5'.
    block (DataFrame): DataFrame containing block-level geographical data.
    tract (DataFrame): DataFrame containing tract-level geographical data.
    zip_code_df (DataFrame): DataFrame containing ZIP code-level geographical data.
    first_names (DataFrame): DataFrame containing probabilities based on first names, including a row for 'ALL OTHER FIRST NAMES'.
    surname_2010_census (DataFrame): DataFrame containing probabilities based on surnames, including a row for 'all other names'.

    Returns:
    DataFrame: A DataFrame with the BISFG probabilities for each racial group.
    """
    
    # Standardize the 'firstname' and 'name' columns for consistency in matching
    input_df['firstname'] = input_df['firstname'].str.upper()
    input_df['name'] = input_df['name'].str.lower().str.replace('-', '_')

    # Copy the input DataFrame to keep the original data intact
    output_df = input_df.copy()

    # Define the suffixes for the racial groups in the surname probability dataset
    race_suffixes = {
        'white': 'white', 'black': 'black', 'aian': 'aian',
        'api': 'api', 'mult_other': '2prace', 'hispanic': 'hispanic'
    }

    # Replace missing or non-existent firstnames with 'ALL OTHER FIRST NAMES'
    output_df['firstname'] = output_df['firstname'].apply(lambda x: x if x in first_names['firstname'].values else 'ALL OTHER FIRST NAMES')

    # Handle hyphenated last names
    def handle_hyphenated_surnames(last_name):
        last_names = last_name.split('_')
        if len(last_names) == 1:
            return last_names[0]
        elif len(last_names) == 2:
            if last_names[0] in surname_2010_census['name'].values:
                return last_names[0]
            elif last_names[1] in surname_2010_census['name'].values:
                return last_names[1]
            else:
                return 'all other names'
        else:
            return 'all other names'

    output_df['name'] = output_df['name'].apply(handle_hyphenated_surnames)

    # Define the function to calculate the BISFG probability for a row
    def bisfg_probability(row, race_key, race_suffix):
        p_r_s = p_f_r = p_g_r = 1

        # Handle hyphenated surnames by splitting and checking each part
        surnames = row['name'].split('_')
        found_surname = False
        for surname in surnames:
            # Check if the surname is in the surname_2010_census dataframe
            matching_surnames = surname_2010_census[surname_2010_census['name'] == surname]
            if not matching_surnames.empty:
                p_r_s *= matching_surnames[f'pct{race_suffix}'].iloc[0]
                found_surname = True

        # Use 'all other names' probabilities if surname is not found
        if not found_surname:
            p_r_s *= surname_2010_census[surname_2010_census['name'] == 'all other names'][f'pct{race_suffix}'].iloc[0]

        # Check for the first name in the first names dataset and get its probability
        firstname = row['firstname']
        matching_firstnames = first_names[first_names['firstname'] == firstname]
        if not matching_firstnames.empty:
            p_f_r = matching_firstnames[f'name_given_{race_key}'].iloc[0]
        else:
            # Use 'ALL OTHER FIRST NAMES' probabilities if firstname is not found
            p_f_r = first_names[first_names['firstname'] == 'ALL OTHER FIRST NAMES'][f'name_given_{race_key}'].iloc[0]

        # Determine the geography-based probability using the block, tract, or ZIP code data
        if 'GeoInd_Block' in input_df.columns and row.get('GeoInd_Block') and row['GeoInd_Block'] in block['GeoInd_Block'].values:
            p_g_r = block[block['GeoInd_Block'] == row['GeoInd_Block']][f'here_given_{race_key}'].iloc[0]
        elif 'GeoInd_Tract' in input_df.columns and row.get('GeoInd_Tract') and row['GeoInd_Tract'] in tract['GeoInd_Tract'].values:
            p_g_r = tract[tract['GeoInd_Tract'] == row['GeoInd_Tract']][f'here_given_{race_key}'].iloc[0]
        elif 'ZCTA5' in input_df.columns and row.get('ZCTA5') and row['ZCTA5'] in zip_code_df['ZCTA5'].values:
            p_g_r = zip_code_df[zip_code_df['ZCTA5'] == row['ZCTA5']][f'here_given_{race_key}'].iloc[0]
        else:
            p_g_r = 1

        # Calculate the numerator of the BISFG probability
        numerator = p_r_s * p_f_r * p_g_r
        
        # Calculate the denominator as the sum of all possible combinations of surname, firstname, and geography probabilities
        denominator = sum(
            p_surname * p_firstname * p_geography
            for key, suffix in race_suffixes.items()
            for part in surnames
            for geo_df, geo_col in [(block, 'GeoInd_Block'), (tract, 'GeoInd_Tract'), (zip_code_df, 'ZCTA5')]
            if part in surname_2010_census['name'].values
            for p_surname in [surname_2010_census[surname_2010_census['name'] == part][f'pct{suffix}'].iloc[0]]
            if firstname in first_names['firstname'].values
            for p_firstname in [first_names[first_names['firstname'] == firstname][f'name_given_{key}'].iloc[0]]
            if row.get(geo_col) in geo_df[geo_col].values
            for p_geography in [geo_df[geo_df[geo_col] == row.get(geo_col)][f'here_given_{key}'].iloc[0]]
        )

        # Return the BISFG probability for the row, handling division by zero
        return numerator / denominator if denominator != 0 else np.nan
       
    # Calculate the BISFG probabilities for each racial group and add them to the output DataFrame
    for key, suffix in race_suffixes.items():
        output_df[f'bisfg_{key}'] = output_df.apply(lambda row: bisfg_probability(row, key, suffix), axis=1)

    return output_df
```
